```python
import pandas as pd
import numpy as np
```

```python
books=pd.read_csv('BX-Books.csv', sep=';',on_bad_lines='skip',encoding='latin-1')
```

```python
books=books.drop(['Image-URL-S','Image-URL-M'], axis=1)
```

```python
users=pd.read_csv('BX-Users.csv', sep=';',on_bad_lines='skip',encoding='latin-1')
```

```python
users.rename(columns={'User-ID':'user_id','Location':'location','Age':'age'}, inplace=True)
```

```python
ratings=pd.read_csv('BX-Book-Ratings.csv', sep=';',on_bad_lines='skip',encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'rating'}, inplace=True)
```

```python
ratings.shape
```

```python
print(books.shape, users.shape, ratings.shape)
```

```python
x=ratings['user_id'].value_counts() >200
```

```python
x[x].shape
```

```python
y=x[x].index
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
rating_with_book=ratings.merge(books, on='ISBN')
```

```python
rating_with_book.head()
```

```python
number_ratings=rating_with_book.groupby('title').count()['rating'].reset_index()
```

```python
number_ratings.head()
```

```python
number_ratings.rename(columns={'rating':'number_of_ratings'}, inplace=True)
```

```python
final_rating=rating_with_book.merge(number_ratings, on='title')
```

```python
final_rating=final_rating[final_rating['number_of_ratings']>=50]
```

```python
final_rating.head()
```

```python
final_rating.head()
```

```python
final_rating.drop_duplicates(['user_id','title'], inplace=True)
```

```python
final_rating.isnull().sum()
```

```python
book_pivot=final_rating.pivot_table(index='title', columns='user_id', values='rating')
```

```python
book_pivot
```

```python

book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
book_pivot.iloc[237,:]
```

```python
distance,suggestion=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
```

TESTING MODEL

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
```

```python
book_name = "2nd Chance"
recommend_book(book_name)
```
